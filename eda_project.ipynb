# Gender Distriubution (Male vs Female)

```python
# Count the occurrences of each gender
gender_counts = df['Gender'].value_counts()

# Function to format the labels with both count and percentage
def func(pct, allvals):
    absolute = int(pct / 100.*sum(allvals))  # Calculate the absolute count
    return f"{absolute} ({pct:.1f}%)"  # Format: count (percentage)

# Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct=lambda pct: func(pct, gender_counts), startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Gender Distribution')
plt.show()

# Print the count of males and females
print(f"Count of Males: {gender_counts['M']}")
print(f"Count of Females: {gender_counts['F']}")
```

# Calculate BMI

```python
# Calculate BMI
df['Height (m)'] = df['Height'] / 100  # Convert height to meters
df['BMI'] = df['Weight'] / df['Height (m)']**2

# Plot BMI distribution
sns.histplot(df['BMI'], kde=True, color='purple', bins=10)
plt.title('BMI Distribution')
plt.show()
```

# Combined histograms for Age and Sleep Duration

```python
# Combined histograms for Age and Sleep Duration
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Age Histogram
sns.histplot(df['Age'], kde=True, color='teal', bins=10, ax=axes[0])
axes[0].set_title('Age Distribution', fontsize=16)
axes[0].set_xlabel('Age', fontsize=14)
axes[0].set_ylabel('Frequency', fontsize=14)

# Sleep Duration Histogram
sns.histplot(df['Sleep duration'], kde=True, color='orange', bins=10, ax=axes[1])
axes[1].set_title('Sleep Duration Distribution', fontsize=16)
axes[1].set_xlabel('Sleep Duration (hours)', fontsize=14)
axes[1].set_ylabel('Frequency', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```
