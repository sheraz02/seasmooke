# Load DataFrame


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data setup
data = pd.read_csv('employee_data.csv')

df = pd.DataFrame(data)
df
```

# Get DataFrame Info

```python
df.info()
```

# Calculate average performance score by department

```python
# Calculate average performance score by department

def avg_performance():
    avg_performance = df.groupby('Department')['PerformanceScore'].mean()

# Plotting
    plt.figure(figsize=(10, 6))
    avg_performance.plot(kind='bar', color='skyblue')
    plt.title('Average Performance Score by Department')
    plt.xlabel('Department')
    plt.ylabel('Average Performance Score')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

avg_performance()
```

# Scatter Plot of Age vs. Salary with Performance Color Coding

```python

def age_vs_salary():
    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(df['Age'], df['Salary'], c=df['PerformanceScore'], cmap='viridis', s=100)
    plt.title('Age vs Salary with Performance Score Color-Coding')
    plt.xlabel('Age')
    plt.ylabel('Salary')
    plt.colorbar(scatter, label='Performance Score')
    plt.tight_layout()
    plt.show()
    
age_vs_salary()
```

# Line Plot of Salary Over Time (Joining Date)


```python
# Convert the DateOfJoining to datetime
def SalaryOverTime():
    df['DateOfJoining'] = pd.to_datetime(df['DateOfJoining'])

    # Sort by Date of Joining
    df_sorted = df.sort_values(by='DateOfJoining')

    plt.figure(figsize=(10, 6))
    plt.plot(df_sorted['DateOfJoining'], df_sorted['Salary'], marker='o', color='green', linestyle='-', markersize=6)
    plt.title('Salary Trend Over Time by Date of Joining')
    plt.xlabel('Date of Joining')
    plt.ylabel('Salary')
    plt.tight_layout()
    plt.xticks(rotation=45)
    plt.show()
SalaryOverTime()
```
